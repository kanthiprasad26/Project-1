```python
import matplotlib.pyplot as plt   ##Used for creating basic charts and graphs
import seaborn as sns             ##  A visualization library that makes graphs look more stylish

# Set style for seaborn
sns.set(style="whitegrid")  ## Makes graphs look clean and readable with a white grid background.

# Top Teams with Most Wins
plt.figure(figsize=(14,7))
wins = df_matches['winner'].value_counts()
wins.plot(kind='bar', color='skyblue')
plt.title("Top Teams with Most Wins", fontsize=14)
plt.xlabel("Teams", fontsize=12)
plt.ylabel("Wins", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
#Toss Decision Trends (Pie Chart) 
toss_decision = df_matches['toss_decision'].value_counts()
plt.figure(figsize=(7,7))
plt.pie(toss_decision, labels=toss_decision.index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue'])
plt.title("Toss Decision Trends", fontsize=14)
plt.show()
```

```python
# Most Man of the Match Awards (Bar Chart) 
plt.figure(figsize=(12,6))
mom_awards = df_matches['player_of_match'].value_counts().head(10)
mom_awards.plot(kind='bar', color='lightgreen')
plt.title("Most Man of the Match Awards", fontsize=14)
plt.xlabel("Players", fontsize=12)
plt.ylabel("Awards", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
#  Runs Distribution per Match (Histogram) 
plt.figure(figsize=(10,6))
sns.histplot(df_matches['result_margin'], bins=20, kde=True, color='orange')
plt.title("Runs Distribution per Match", fontsize=14)
plt.xlabel("Win by Runs", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
# Win Margin Analysis (Box Plot) 
plt.figure(figsize=(10,6))
sns.boxplot(x=df_matches['result_margin'], color='purple')
plt.title("Win Margin Analysis", fontsize=14)
plt.xlabel("Win by Runs", fontsize=12)
plt.show()
```

```python
# IPL Winners Over the Years (barchart) 
plt.figure(figsize=(12,6))
winners_per_year = df_matches.groupby("season")['winner'].value_counts().unstack().T
winners_per_year.plot(kind='bar', figsize=(12,6))
plt.title("IPL Winners Over the Years", fontsize=14)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Wins", fontsize=12)
plt.xticks(rotation=45)
plt.show()

```

```python
# Set seaborn style for better visualization
sns.set(style="whitegrid")

# Create figure
plt.figure(figsize=(12,6))

# Group data by season and count winners
winners_per_year = df_matches.groupby("season")['winner'].value_counts().unstack().T

# Plot horizontal bar chart
winners_per_year.plot(kind='barh', figsize=(12,6), colormap='Paired')

# Add labels and title
plt.title("IPL Winners Over the Years", fontsize=14)
plt.xlabel("Wins", fontsize=12)
plt.ylabel("Teams", fontsize=12)

# Rotate y-axis labels for better readability
plt.yticks(rotation=0)

# Show the plot
plt.show()
```

```python
# Head-to-Head Wins (Stacked Bar Chart) 
rival_teams = ["Mumbai Indians", "Chennai Super Kings"]
h2h = df_matches[df_matches['team1'].isin(rival_teams) & df_matches['team2'].isin(rival_teams)]
h2h_wins = h2h['winner'].value_counts()
h2h_wins.plot(kind='bar', color=['blue', 'yellow'])
plt.title("MI vs CSK Head-to-Head Wins", fontsize=14)
plt.ylabel("Wins")
plt.xticks(rotation=0)
plt.show()

```

```python
#  Top Run Scorers Over Seasons (Scatter Plot) 

batsmen = deliver_df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.scatterplot(x=batsmen.index, y=batsmen.values, s=100, color='red')
plt.title("Top Run Scorers Over Seasons", fontsize=14)
plt.xlabel("Batsmen", fontsize=12)
plt.ylabel("Total Runs", fontsize=12)
plt.xticks(rotation=45)
plt.show()

##PUT Leagend by year 
```

```python
## Top Wicket-Takers
wicket_deliveries = deliver_df[deliver_df['is_wicket'] == 1]

# Count wickets for each bowler
wicket_counts = wicket_deliveries['bowler'].value_counts().head(10)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=wicket_counts.values, y=wicket_counts.index, palette='viridis')
plt.title("Top 10 Wicket-Takers in IPL", fontsize=14)
plt.xlabel("Total Wickets", fontsize=12)
plt.ylabel("Bowler", fontsize=12)
plt.show()
```

```python
## **Top Wicket-Takers

# Count wickets for each bowler
wicket_counts = wicket_deliveries['bowler'].value_counts().head(10)

# Plot without palette OR use hue=y
plt.figure(figsize=(12,6))
sns.barplot(x=wicket_counts.values, y=wicket_counts.index, hue=wicket_counts.index, legend=False)
plt.title("Top 10 Wicket-Takers in IPL", fontsize=14)
plt.xlabel("Total Wickets", fontsize=12)
plt.ylabel("Bowler", fontsize=12)
plt.show()
```

```python
#  Home vs. Away Wins (Grouped Bar Chart) 
home_away = df_matches.groupby(['venue', 'winner']).size().unstack().sum().sort_values(ascending=False)
home_away.plot(kind='bar', figsize=(12,6), color='teal')
plt.title("Home vs Away Wins", fontsize=14)
plt.xlabel("Teams", fontsize=12)
plt.ylabel("Wins", fontsize=12)
plt.xticks(rotation=45)
plt.show()

```

```python
# Most Common Match Venues (Bar Chart) 
plt.figure(figsize=(12,6))
venues = df_matches['venue'].value_counts().head(10)
venues.plot(kind='bar', color='brown')
plt.title("Most Common Match Venues", fontsize=14)
plt.xlabel("Venues", fontsize=12)
plt.ylabel("Matches Played", fontsize=12)
plt.xticks(rotation=45)
plt.show()

```
